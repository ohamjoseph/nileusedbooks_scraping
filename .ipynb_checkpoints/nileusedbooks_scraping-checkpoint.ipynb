```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
n = 1

links = {}
links['liens'] = []

for i in range(1,n+1):
    print("----------------")
    url = 'https://nileusedbooks.com/product-category/french/page/'+str(i)
    response = requests.get(url)
    html = response.content
    soup = bs(html, "html.parser")
    divs = soup.find_all("div", class_="item-img products-thumb")
    for div in divs:
        hrf = div.find_all("a")[0]['href'].strip()
        print(hrf)
        links['liens'].append(hrf)
```

```python
df = pd.DataFrame(links)
df = df.drop_duplicates()
df
```

```python
donnee = df.iloc[:4,]
donnee
```

```python
url = 'https://nileusedbooks.com/product/'
donne = {}
donne['Book Title'] = []
donne['ISBN Number'] = []
donne['Price in Egyptian Pounds'] = []
donne['Price in USD'] = []

for n in donnee.liens.values:
    url = 'https://nileusedbooks.com/product/'+n
    response = requests.get(url)
    html = response.content
    soup = bs(html, "html.parser")
    prix = int(soup.find_all("ins")[0].find('bdi').text.split()[0][:-3])
    nbd = soup.find_all('div', class_='tab-content')[0].div.find_all('p')[1].text
    print(prix)
    donne['Book Title'].append(n)
    donne['ISBN Number'].append(nbd)
    donne['Price in Egyptian Pounds'].append(prix)
    donne['Price in USD'].append(prix/19)
    
    
```

```python
donne_df = pd.DataFrame(donne)
donne_df
```

```python
donne_df.to_excel('file.xlsx', index=False)
```
